```python
UMBRAL_URBANO = df['vlr_ingr_salario'].quantile(0.75)
```

```python
UMBRALMEDIO_URBANO = df['vlr_ingr_salario'].quantile(0.5)
```

```python
def asignar_zona(row):
    salario = row['vlr_ingr_salario']
    ingr_agricola = row['vlr_ingr_cosecha']
    emp = row['tip_empleado']

    # 1. Agrícola
    if ingr_agricola > 0 or emp in [7, 10]:
        return 'Agrícola'

    # 2. Urbana
    if emp in [2, 4, 5]:  # Ajusta el umbral a lo que consideres alto
        return 'Urbana'

    # 3. Mineral
    if emp == 1 and salario > UMBRAL_URBANO:
        return 'Mineral'

    # 4. Avícola
    if emp in [6, 8, 9]:
        return 'Avícola'

    # 5. No clasificados
    return np.nan

# Aplicar al DataFrame
df['zona_final'] = df.apply(asignar_zona, axis=1)
```

```python
print(df['zona_final'].value_counts())
```

```python
import pandas as pd
from tabulate import tabulate  # opcional para imprimir bonito en consola

# Función para construir el perfil por zona con tus variables corregidas
def construir_perfil(df):
    perfil = df.groupby('zona_final').agg(
        Total=('zona_final', 'count'),
        Edad_Promedio=('edad_calculada', 'mean'),
        Porc_Mujeres=('sexo_persona', lambda x: (x == 2).mean() * 100),
        Porc_Educ_Baja=('niv_educativo', lambda x: (x.isin([0, 1, 2])).mean() * 100),
        Porc_Pobreza_Multidimensional=('C', lambda x: (x > 5).mean() * 100),
        Salario_Promedio=('vlr_ingr_salario', 'mean')
    ).round(2)

    # Formatear columnas para presentación
    perfil['Edad_Promedio'] = perfil['Edad_Promedio'].astype(int)
    perfil['Salario_Promedio'] = perfil['Salario_Promedio'].apply(lambda x: f"${x:,.0f}")
    perfil['Porc_Mujeres'] = perfil['Porc_Mujeres'].astype(str) + "%"
    perfil['Porc_Educ_Baja'] = perfil['Porc_Educ_Baja'].astype(str) + "%"
    perfil['Porc_Pobreza_Multidimensional'] = perfil['Porc_Pobreza_Multidimensional'].astype(str) + "%"

    return perfil

# Ejecutar función
tabla_perfil = construir_perfil(df)

# Mostrar bonito en Jupyter
display(tabla_perfil)

# Mostrar bonito en consola
print(tabulate(tabla_perfil, headers='keys', tablefmt='fancy_grid'))
```
