1. Los diccionarios tienen una clave unica, y un valor asociado a la clave (string)
2. No se pueden indezar (No se sabe la pocision de datos)
3. Son mutables

````python
diccionario={"Enero":22,"Febrero":15,"Marzo":71}
type(diccionario)
````

````python
#Convertir diccionario a dataframe
ventas1=pd.Series(diccionario)
ventas1
````

````python
alex=pd.Series(diccionario,index=["Enero","Febrero","Marzo","Abril"])
alex
````

````python
medidas=pd.DataFrame(
    {"Humedad":[41,20,30,80],
     "Salinidad":[22,15,30,80],
     "Sensórica":["Sí","No","No","No"],
     "Acidez":["alta","media","baja","baja"],
     "ph":[1,7,14,13]},
    index=["La Ureña","Paso Chino","La Laguna","Los Teres"],
)

medidas
````
